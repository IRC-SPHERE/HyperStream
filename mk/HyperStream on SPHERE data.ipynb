```python
from hyperstream import *
from hyperstream_bases import *
from hyperstream_modifiers import *
from hyperstream_intervals import *
from hyperstream_sphere import *
import pprint as pp
class AttrDict(dict):
  def __getattr__(self,key):
    return(self.__getitem__(key))
  def __setattr__(self,key,value):
    self.__setitem__(key,value)
T = ToolBase(1,'codebase/tools',MAX_DATE)
locals().update(T.versions)
M = MemoryBase(2)
S = SphereBase(3,date(2016,7,13,0,0,0,0,UTC))
t1 = date(2016,4,28,20,0,0,0,UTC)
t2 = date(2016,4,29,13,0,0,0,UTC)
e = 'environmental'
v = 'video'
uid_list = S[e,t1,t1+HOUR,Component('uid')+Data()+List()]()
uid_set = set(uid_list)
pp.pprint(uid_set)
rooms = set()
sensor_types = set()
uid_info = {}
for uid in uid_set:
  info = AttrDict()
  uid_info[uid] = info
  match = re.match('^(.*)_(.*)$',uid)
  if match:
    (info.sensor_type,info.room) = match.groups()
  else:
    (info.sensor_type,info.room) = (u'EL',None)
  rooms.add(info.room)
  sensor_types.add(info.sensor_type)
  info.example = S[e,t1,t1+MINUTE,CompFilter('uid',[uid])+DelNones()+Data()+Head(3)+List()]()
pp.pprint(rooms)
pp.pprint(sensor_types)
pp.pprint(uid_info)
room_filter = AttrDict()
for room in rooms:
  room_filter[room] = CompFilter('uid',[uid for uid in uid_set if uid_info[uid].room==room])
res = S[e,t1,t1+MINUTE,room_filter.K+Data()+List()]()
res = S[e,t1,t1+HOUR,room_filter.K+DelNones()+Component('temperature-S1_K')+Data()+Average()]()
res = S[e,t1,t1+3*MINUTE,Component('motion-S1_K')+Data()+List()]()
M['every10s'] = clock(stride=10*SEC)
M['every_min'] = clock(stride=MINUTE)
M['x'] = pool(timer=M['every10s'],data=S[e,-MINUTE,delta(0),Component('motion-S1_K')],func=Data()+Count())
M['x'] = pool(timer=M['every10s'],data=S[e,-MINUTE,delta(0),Component('motion-S1_K')],func=Data()+List())
res = M['x',t1,t1+5*MINUTE,Data()+List()]()
pp.pprint(res)
```
